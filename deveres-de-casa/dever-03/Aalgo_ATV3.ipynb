```python
def eh_palindromo(array, inicio=0, fim=None):
    """
    Verifica recursivamente se um array é um palíndromo.

    Um array é palíndromo quando pode ser lido da mesma forma
    da esquerda para a direita e da direita para a esquerda.
    """
    if fim is None:
        fim = len(array) - 1

    # Caso base: índices se cruzaram ou são iguais
    if inicio >= fim:
        return True

    # Se os elementos forem diferentes, não é palíndromo
    if array[inicio] != array[fim]:
        return False

    # Chamada recursiva
    return eh_palindromo(array, inicio + 1, fim - 1)


def ler_vetor():
    """
    Permite ao usuário definir o tamanho do vetor
    e inserir seus elementos.
    """
    tamanho = int(input("Quantos elementos terá o vetor? "))

    vetor = []

    for i in range(tamanho):
        elemento = input(f"Digite o elemento {i + 1}: ")
        vetor.append(elemento)

    return vetor


def main():
    """Função principal do programa."""
    vetor = ler_vetor()

    if eh_palindromo(vetor):
        print("O vetor é palíndromo.")
    else:
        print("O vetor NÃO é palíndromo.")


if __name__ == "__main__":
    main()
```
