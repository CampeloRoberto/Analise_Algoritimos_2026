```python
def calcular_f(n: int) -> int:
    """
    Calcula o valor da função recursiva:
        F(n) = 2 * F(n-1) + n^2
    com caso base F(1) = 2.

    Args:
        n (int): valor de entrada (n >= 1)

    Returns:
        int: resultado da função F(n)
    """
    if n < 1:
        raise ValueError("n deve ser maior ou igual a 1")

    if n == 1:
        return 2

    return 2 * calcular_f(n - 1) + n**2


def main() -> None:
    """Solicita um valor ao usuário e exibe F(n)."""
    try:
        n = int(input("Digite o valor de n (>= 1): "))
        resultado = calcular_f(n)
        print(f"F({n}) = {resultado}")
    except ValueError as erro:
        print(f"Erro: {erro}")


if __name__ == "__main__":
    main()
```
