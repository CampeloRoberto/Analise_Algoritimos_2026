```python
"""
Recursive Factorial Algorithm - Execution Time Measurement

This program calculates the factorial of a number using recursion
and measures the execution time of the algorithm for the value
provided by the user.
"""

import time
import sys

# Increase recursion limit to allow larger factorial calculations
sys.setrecursionlimit(5000)


def calculate_factorial(n):
    """
    Compute the factorial of an integer using recursion.

    The factorial of n is defined as:
        n! = n * (n - 1)!

    Base cases:
        0! = 1
        1! = 1

    Args:
        n (int): Non-negative integer.

    Returns:
        int: Factorial result.
    """
    """
Factorial Recursive Algorithm - Execution Time Test
-------------------------------------------------------
n = 10
Execution time: 0.00000525 seconds
-------------------------------------------------------
n = 100
Execution time: 0.00002670 seconds
-------------------------------------------------------
n = 500
Execution time: 0.00027418 seconds
-------------------------------------------------------
n = 1000
Execution time: 0.00065637 seconds

Os resultados obtidos mostram que o tempo de execução do algoritmo aumenta conforme o valor de n cresce.
Para n = 10, o tempo registrado é extremamente pequeno, enquanto para n = 100 já se observa um pequeno aumento.
Quando o valor passa para n = 500 e n = 1000, o tempo continua aumentando gradualmente.
Esse comportamento ocorre porque o algoritmo de fatorial recursivo executa uma nova chamada da função para cada valor até chegar ao caso base.
Assim, quanto maior for o número informado, maior será a quantidade de chamadas realizadas e, consequentemente,
maior será o tempo necessário para concluir o cálculo.

Para determinar a complexidade assintótica, analisamos quantas operações o algoritmo executa em função de n, e não apenas o tempo medido.
No caso do fatorial recursivo, cada chamada da função executa um conjunto pequeno e constante de operações:
uma verificação do caso base, uma multiplicação e uma nova chamada recursiva.
Podemos representar a quantidade de operações da seguinte forma:
T(n) = T(n - 1) + c
Sendo:
T(n) = tempo para calcular o fatorial de n;
T(n - 1) = tempo da chamada recursiva seguinte;
c = quantidade constante de operações executadas em cada chamada;
Expandindo a relação:
T(n) = T(n - 1) + c
T(n) = (T(n - 2) + c) + c
T(n) = T(n - 2) + 2c
T(n) = T(n - 3) + 3c
...
T(n) = T(1) + (n - 1)c
Como T(1) é constante, o termo dominante passa a ser:
T(n) ≈ n · c
Ignorando constantes (regra da notação assintótica):
T(n) ∈ O(n)
"""

    # Base case: factorial of 0 or 1 is 1
    if n == 0 or n == 1:
        return 1

    # Recursive call
    return n * calculate_factorial(n - 1)


def main():
    """
    Main program execution.

    The program asks the user for a number, calculates its factorial
    using recursion, and measures the execution time.
    """

    # Ask the user for input
    number = int(input("Enter a number to calculate its factorial: "))

    # Start time measurement
    start_time = time.time()

    # Execute factorial calculation
    result = calculate_factorial(number)

    # End time measurement
    end_time = time.time()

    execution_time = end_time - start_time

    # Display results
    print(f"\nFactorial of {number} is {result}")
    print(f"Execution time: {execution_time:.8f} seconds")


# Standard Python entry point
if __name__ == "__main__":
    main()
```
