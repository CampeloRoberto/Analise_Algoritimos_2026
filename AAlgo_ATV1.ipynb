```python
import random
import time


LIST_SIZES = [1000, 5000, 10000, 20000, 50000]


def insertion_sort(lista):
    """
    Ordena uma lista utilizando o algoritmo Insertion Sort.

    O algoritmo percorre a lista e insere cada elemento na posição
    correta dentro da parte já ordenada.

    Complexidade:
        - Tempo médio: O(n²)
        - Pior caso: O(n²)
        - Melhor caso: O(n)

    Args:
        lista (list): Lista de números a ser ordenada.

    Returns:
        list: Lista ordenada.
    """

    for indice_atual in range(1, len(lista)):
        valor_atual = lista[indice_atual]
        indice_anterior = indice_atual - 1

        while indice_anterior >= 0 and lista[indice_anterior] > valor_atual:
            lista[indice_anterior + 1] = lista[indice_anterior]
            indice_anterior -= 1

        lista[indice_anterior + 1] = valor_atual

    return lista


def gerar_lista_aleatoria(tamanho):
    """
    Gera uma lista de números aleatórios.

    Args:
        tamanho (int): Quantidade de elementos da lista.

    Returns:
        list: Lista contendo números aleatórios.
    """
    return [random.randint(0, 100000) for _ in range(tamanho)]


def medir_tempo_insertion_sort(lista):
    """
    Mede o tempo de execução do Insertion Sort.

    Args:
        lista (list): Lista de números.

    Returns:
        float: Tempo de execução em segundos.
    """
    inicio = time.time()
    insertion_sort(lista)
    fim = time.time()

    return fim - inicio


def medir_tempo_sorted(lista):
    """
    Mede o tempo de execução do sorted() (Timsort).

    Args:
        lista (list): Lista de números.

    Returns:
        float: Tempo de execução em segundos.
    """
    inicio = time.time()
    sorted(lista)
    fim = time.time()

    return fim - inicio


def executar_experimento():
    """
    Executa o experimento comparando Insertion Sort e sorted().

    Gera listas aleatórias de diferentes tamanhos e mede
    o tempo de execução de cada algoritmo.
    """

    print("Comparação entre Insertion Sort O(n²) e sorted() O(n log n)")
    print("-" * 60)

    for tamanho in LIST_SIZES:

        lista_original = gerar_lista_aleatoria(tamanho)

        lista_insertion = lista_original.copy()
        lista_sorted = lista_original.copy()

        tempo_insertion = medir_tempo_insertion_sort(lista_insertion)
        tempo_sorted = medir_tempo_sorted(lista_sorted)

        print(f"Tamanho da lista: {tamanho}")
        print(f"Insertion Sort: {tempo_insertion:.4f} segundos")
        print(f"sorted() (Timsort): {tempo_sorted:.4f} segundos")
        print("-" * 60)


if __name__ == "__main__":
    executar_experimento()
```
